```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x=np.array([0,10,20,30])
y=np.array([4,40,34,55])
```

```python
plt.plot(x,y)
```

```python
plt.show()
```

```python
plt.plot(x,y,"*")
```

```python
plt.plot(x,y,marker="*",ms=20)
```

```python
plt.plot(x,y,'*:g')
plt.grid()
```

```python
plt.scatter(x,y,color="hotpink")
```

```python
x=np.array([2,3,4,6])
y=np.array([23,24,34,38])
plt.subplot(1,2,1)
plt.plot(x,y,'*:g')

x=np.array([1,3,5,7])
y=np.array([22,32,26,44])
plt.subplot(1,2,2)
plt.plot(x,y,'*:b')
```
