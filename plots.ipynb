```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
import pandas as pd
```

```python
x=np.array([3,4,5,7])
mylabels=["Tea","Coffee","Cold drink","Cold Coffee"]
myexplode=[0,0.1,0.2,0.3]
plt.pie(x,labels=mylabels)
plt.show()
```

```python
plt.pie(x,labels=mylabels,explode=myexplode)
plt.show()
```

```python
y=np.array([45,55,40,42])
plt.bar(x,y,color="b")
plt.show
```

```python
y=np.random.normal(200,20,150)
plt.hist(y,color="r")
plt.show()
```

```python
x=np.array([2,3,4,5,6,7,8])
y=np.array([22,33,44,55,23,20,34])
plt.scatter(x,y,color="hotpink")
plt.show()
```
